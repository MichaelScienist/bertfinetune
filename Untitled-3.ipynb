```python
import pandas as pd

# get all data set 
original_df = pd.read_csv('test_dataset.csv')

epoch0 = pd.read_csv('p0.csv')
epoche = pd.read_csv('p3.csv')
epoch10 = pd.read_csv('p10.csv')
original_df['sentiment'] = original_df['sentiment'].map({'positive': 1, 'negative': 0})
```

```python
epoch0
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

#plot the bar chart  
accuracy_0 = accuracy_score(original_df['sentiment'], epoch0['predictions'])
accuracy_3 = accuracy_score(original_df['sentiment'], epoche['predictions'])
accuracy_10 = accuracy_score(original_df['sentiment'], epoch10['predictions'])

comparison_df = pd.DataFrame({
    'Epochs': ['0', '3', '10'],
    'Accuracy': [accuracy_0, accuracy_3, accuracy_10]
})
#plot the bar chart
plt.figure(figsize=(8, 6))
plt.bar(comparison_df['Epochs'], comparison_df['Accuracy'], color=['red', 'green', 'blue'])
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Comparison of Model Accuracy Across Different Epochs')
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# get all th escore
precision_0 = precision_score(original_df['sentiment'], epoch0['predictions'])
recall_0 = recall_score(original_df['sentiment'], epoch0['predictions'])
f1_0 = f1_score(original_df['sentiment'], epoch0['predictions'])

precision_3 = precision_score(original_df['sentiment'], epoche['predictions'])
recall_3 = recall_score(original_df['sentiment'], epoche['predictions'])
f1_3 = f1_score(original_df['sentiment'], epoche['predictions'])

precision_10 = precision_score(original_df['sentiment'], epoch10['predictions'])
recall_10 = recall_score(original_df['sentiment'], epoch10['predictions'])
f1_10 = f1_score(original_df['sentiment'], epoch10['predictions'])

# make a matrix
metrics_df = pd.DataFrame({
    'Epochs': ['0', '3', '10'],
    'Precision': [precision_0, precision_3, precision_10],
    'Recall': [recall_0, recall_3, recall_10],
    'F1 Score': [f1_0, f1_3, f1_10]
})

# plot
plt.figure(figsize=(12, 8))
for i, metric in enumerate(['Precision', 'Recall', 'F1 Score']):
    plt.subplot(1, 3, i+1)
    plt.bar(metrics_df['Epochs'], metrics_df[metric], color=['red', 'green', 'blue'])
    plt.xlabel('Epochs')
    plt.ylabel(metric)
    plt.title(f'Comparison of {metric} Across Epochs')

plt.tight_layout()
plt.show()
```
